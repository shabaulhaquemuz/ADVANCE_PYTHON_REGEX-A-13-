```python
# Data point: A data point is a single pair of values on a graph — one from the x-axis and one from the y-axis.
# Example: (2, 10) is a data point.

# The x-axis is the horizontal line on a graph.
# It usually represents the independent variable (the input)
# The y-axis is the vertical line on a graph.
# It usually represents the dependent variable (the output)

# Figure: The entire window or image where plots are drawn.
# It can contain one or more plots (axes).

# Axes: The area within a figure where data is plotted (includes x and y axis).
# Think of it as a single plot inside the figure.

# Plot: A visual representation of data (like a line, bar, or scatter).
# Made using functions like plt.plot(), plt.bar(), etc.

# Marker: A symbol that marks each data point on the plot.
# Example: 'o' for circle, '*' for star.

# Line style: The style of the line connecting points.
# Example: '-' for solid, '--' for dashed.

# Color: Defines the color of lines, markers, etc. in the plot.
# Example: 'r' for red, 'b' for blue.

# Legend: Explains what different lines or markers in the plot represent.
# Used when there are multiple plots in one graph.

# Label: Text used to describe axes or data series.
# Commonly set using xlabel(), ylabel(), and label= in plot().
```

```python
import matplotlib.pyplot as plt
x=["Mon","Tue","Wed","Thur","Fri","Sat","Sun"]
y=[3,17,5,6,10,15,2]
plt.plot(x,y)
plt.xlabel("day of the week")
plt.ylabel("sales per day")
plt.title("BAKERY SALES THIS WEEK")
plt.show()
```

```python
# Matplotlib supports two main ways to create plots:

# Pyplot style (like MATLAB)
# Object-Oriented Approach (OOA) — more flexible and professional

# The Object-Oriented Approach in Matplotlib means creating plots by first making a Figure and Axes object, and then calling methods on them to 
# build the plot.
# It gives better control and is preferred for complex or multiple plots. an Axes object is the actual area where data is plotted — it contains the x-axis,
# y-axis, lines, labels, and titles.

# | Feature                    | **Pyplot Style (`plt`)**                    | **Object-Oriented Approach (OOA)**                    |
# | -------------------------- | ------------------------------------------- | ----------------------------------------------------- |
# | **Use Case**               | For single, simple projects                 | For large, complex projects                           |
# | **Flexibility**            | Less flexible                               | More flexible                                         |
# | **User Level**             | Good for beginners                          | For becoming a better data scientist                  |
# | **Function Usage**         | Uses high-level functions like `plt.plot()` | Uses `Figure` and `Axes` objects explicitly           |
# | **Ease of Coding**         | Easier and quicker for simple plots         | Slightly more code, but better structure              |
# | **Figure & Axes Handling** | Done automatically in the background        | You define them manually (`fig, ax = plt.subplots()`) |
# | **Code Style**             | Procedural                                  | Object-Oriented                                       |
```

```python
import matplotlib.pyplot as plt
months=[1,2,3,4]
sales=[1000,1500,1200,1800]
plt.plot(months,sales,color="orange",linestyle="--",linewidth=2,marker="o",label="Sales per Month")  #The label just stores the description internally.
plt.legend()  #To display it on the plot, you must add:plt.legend()
plt.xlabel("Months")
plt.ylabel("Sales per month")
plt.title("MONTHLY SALES DATA REPORT")
plt.show()
```

```python
import matplotlib.pyplot as plt
months=[1,2,3,4]
sales=[1000,1500,1200,1800]
plt.plot(months,sales,color="orange",linestyle="--",linewidth=2,marker="o",label="Sales per Month")  #The label just stores the description internally.
plt.legend(loc="upper left",fontsize=12)  #upper left/upper right/lower left/ lower right-WE CAN ALSO PASS PARAMETERS IN IT, BUT ITS OPTIONAL,  #To display it on the plot, you must add:plt.legend()
plt.xlabel("Months")
plt.grid(color="grey",linestyle=":",linewidth=1)   #optional parameters can be passed
plt.ylabel("Sales per month")
plt.title("MONTHLY SALES DATA REPORT")
plt.show()
```

```python
#xlim(): used to control the ranges of x-axis and y-axis
import matplotlib.pyplot as plt
months=[1,2,3,4]
sales=[1000,1500,1200,1800]
plt.plot(months,sales,color="orange",linestyle="--",linewidth=2,marker="o",label="Sales per Month")  #The label just stores the description internally.
plt.legend(loc="upper left",fontsize=12)  #upper left/upper right/lower left/ lower right-WE CAN ALSO PASS PARAMETERS IN IT, BUT ITS OPTIONAL,  #To display it on the plot, you must add:plt.legend()
plt.xlabel("Months")
plt.grid(color="grey",linestyle=":",linewidth=1)   #optional parameters can be passed
plt.ylabel("Sales per month")
plt.title("MONTHLY SALES DATA REPORT")
plt.xlim(1,4)
plt.ylim(0,2000)
plt.show()
```

```python
#BAR CHART/ BAR GRAPH:
# ✅ You can use bar chart or bar graph — both are correct.
# But in programming and analytics, "bar chart" is more standard.

# A bar chart is a type of graph that uses rectangular bars to represent data.
# The length or height of each bar shows the value or frequency of the corresponding category.

# Used for comparing categories or groups.
# Bars can be vertical (plt.bar()) or horizontal (plt.barh()).
# Best suited for categorical or discrete data.
```

```python
import matplotlib.pyplot as plt
product=["A","B","C","D"]
sales=[1000,1500,800,1200]
plt.bar(product,sales,color="orange",label="Sales 2025")
plt.legend()
plt.xlabel("Name of Product")
plt.ylabel("sales")
plt.title("Product Sale Comparison")
plt.show()
```

```python
#horizontal bar chart
import matplotlib.pyplot as plt
product=["A","B","C","D"]
sales=[1000,1500,800,1200]
plt.barh(product,sales,color="orange",label="Sales 2025")
plt.legend()
plt.ylabel("Name of Product")
plt.xlabel("sales")
plt.title("Product Sale Comparison")
plt.show()
```

```python
#PIE CHART
# A pie chart is a circular chart divided into slices to show the proportions or percentages of different categories in a dataset.
# The whole circle represents 100% of the data.
# Each slice (wedge) represents a part of the whole.

# A pie chart is used for categorical data where each category represents a part of a whole.


#autopct is a parameter in plt.pie() that displays percentage values on the slices of the pie chart.
#  Format Meaning:
# '%1.1f%%'
# → One decimal place + % sign (e.g., 25.0%)

# '%1.0f%%'
# → No decimal place (e.g., 25%)

# '%1.2f%%'
# → Two decimal places (e.g., 25.00%)
```

```python
import matplotlib.pyplot as plt
regions=["North","South","East","West"]
revenue=[10000,20000,30000,40000]
plt.pie(revenue,labels=regions,autopct="%1.2f%%",colors=["gold","green","magenta","orange"])
plt.show()
```

```python
# A histogram is a graphical representation of continuous numerical data, where the data is divided into ranges (called bins),
# and bars are used to show how many data points fall into each range.

# Used for continuous (not categorical) data.
# The x-axis shows data ranges (bins).
# The y-axis shows frequency (count of values in each bin).
```

```python
#plt.hist(data,bins=range_value,color="coloe_name",edgecolor="black")
import matplotlib.pyplot as plt
scores=[40,45,56,59,65,67,75,79,80,84,89,97,68,78,88]
plt.hist(scores,bins=5,color="purple",edgecolor="black")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.title("Score distribution of Students")
plt.show()
```
