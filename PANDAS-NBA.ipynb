```python
nba["Salary"] = nba["Salary"].astype(int)  
#Cannot convert non-finite values (NA or inf) to integer...You're getting this error because your Salary column contains NaN (missing values).
#Python cannot directly convert NaN (Not a Number) to int.
```

```python
#4.36+e6 means 4.36*10**6, which is equals to 4360000
```

```python
nba["Salary"] = nba["Salary"].fillna(0).astype(int)
print(nba["Salary"])
```

```python
nba["Salary"] = nba["Salary"].astype("Int64")  # Pandas nullable integer
print(nba["Salary"])



# | Situation            | Solution                |
# | -------------------- | ----------------------- |
# | You don't need NaN   | `fillna(0).astype(int)` |
# | You need to keep NaN | `astype("Int64")`       |
```
