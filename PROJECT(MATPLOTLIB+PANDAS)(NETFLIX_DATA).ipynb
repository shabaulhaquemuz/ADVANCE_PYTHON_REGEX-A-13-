```python
### ðŸ“š Common Python Libraries & Functions for Data Visualization

# | Library                 | Purpose                                        |
# |------------------------|------------------------------------------------|
# | pandas                 | Data loading, cleaning, exploration            |
# | matplotlib.pyplot      | Plotting all types of graphs                   |
# | plt.plot()             | Line chart                                     |
# | plt.bar(), plt.barh()  | Bar charts (vertical/horizontal)               |
# | plt.pie()              | Pie chart                                      |
# | plt.hist()             | Histogram (distribution check)                 |
# | plt.scatter()          | Scatter plot                                   |
# | plt.subplot() / plt.subplots() | Multiple plots (subplots)              |
# | plt.tight_layout()     | Fix overlapping labels                         |
# | plt.savefig()          | Save charts as images                          |
```

### ðŸ§­ Data Analysis Workflow Overview

```python
# | Step Name                     | What We'll Do                                                                  |
# |------------------------------|---------------------------------------------------------------------------------|
# | Load the data                | Read the CSV file using pandas.                                                |
# | Clean the data               | Handle missing values, remove duplicates, fix columns if needed.               |
# | Understand the data          | Use head(), info(), describe() to explore the dataset.                         |
# | Identify questions to answer | What do we want to know? (E.g., Movies vs TV Shows, content ratings, etc.)     |
# | Visualize the data (EDA)     | Use Matplotlib to draw charts and answer these questions visually.             |
# | Save the plots               | Use plt.savefig() for your reports.                                            |
```

```python
#step-1 import the libraries
import pandas as pd
import matplotlib.pyplot as plt

#load the data
df=pd.read_csv("netflix_titles.csv")
print(df.head())
#clean data
df=df.dropna(subset=["type","release_year","rating","country","duration"])
```

# ðŸ“Š Visualization Plan Using Matplotlib

```python



# | Question                                                                 | Chart Type                | Matplotlib Function              |
# |--------------------------------------------------------------------------|---------------------------|----------------------------------|
# | How many Movies vs TV Shows?                                            | Bar Chart                 | plt.bar()                        |
# | What is the percentage of each content rating (PG, R, TV-MA)?           | Pie Chart                 | plt.pie()                        |
# | How has the number of releases changed over the years?                  | Line Plot                 | plt.plot()                       |
# | What is the distribution of movie durations?                            | Histogram                 | plt.hist()                       |
# | Relationship between release year and number of shows?                  | Scatter Plot (optional)   | plt.scatter()                    |
# | Top 10 countries with the highest number of shows?                      | Bar Chart( horizontal)    | plt.barh()                       |                                   |
# | Compare multiple plots together (e.g., Movies vs TV Shows by Year)     | Subplots                  | plt.subplot() /plt.subplots()    |
# | Save final charts                                                       | Saving Figures            | plt.savefig()                    |


# ðŸŽ¬ Content Rating Full Forms
# PG â†’ Parental Guidance Suggested
# (Some material may not be suitable for children.)

# R â†’ Restricted
# (Under 17 requires accompanying parent or adult guardian.)

# TV-MA â†’ Television - Mature Audience
# (Intended for adults; may be unsuitable for children under 17.)
```

```python
#Number of Movies vs number of TV shows
type_counts=df["type"].value_counts()
print(type_counts)
print("index is:",type_counts.index)
print("values are:",type_counts.values)
plt.figure(figsize=(10,5))
plt.bar(type_counts.index,type_counts.values,color=["mediumslateblue","thistle"])
plt.xlabel("Type of content")
plt.ylabel("Number of content")
plt.title("Number of movies vs TV shows on Netflix")
plt.tight_layout()
plt.savefig("movies_vs_tvshows_netflix_bargraph.png")
plt.show()
```

```python
#to check distribution of content's rating
rating_counts=df["rating"].value_counts()
print(rating_counts)
plt.figure(figsize=(10,8))
plt.pie(rating_counts,labels=rating_counts.index,autopct="%1.0f%%")
plt.title("Percentage of Content Ratings")
plt.tight_layout()
plt.savefig("content_rating_percentage_piechart.png")
plt.show()
```

```python
#How has the number of releases changed over the years?  
release_per_year=df["release_year"].value_counts().sort_index()
print(release_per_year)
plt.figure(figsize=(10,5))
plt.plot(release_per_year.index,release_per_year.values,color="orange")
plt.title("Content Released Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Contents released")
plt.tight_layout()
plt.savefig("Content_released_per_year_comparison.png")
plt.show()
```

```python
#check distribution of movies duration using histogram
movie_df=df[df["type"]=="Movie"].copy()    #This selects only the rows where "type" is "Movie".
movie_df["duration_int"]=movie_df["duration"].str.replace("min","").astype(int)       #  .str.replace("min", "")---to--.astype(int) i.e. "90 " â†’ 90 (as an integer)
plt.figure(figsize=(8,6))                                                             #i.e. "90 min"  â†’  "90 "  â†’  90
plt.hist(movie_df["duration_int"],bins=30,color="purple",edgecolor="black")
plt.title("Distribution of Movie Duration")
plt.xlabel("Duration(minutes)")
plt.ylabel("Number of movies")
plt.tight_layout()
plt.savefig("Movies_duration_distribution_Histogram.png")
plt.show()

```

```python
#release year vs number of shows
release_year_counts=df["release_year"].value_counts().sort_index() #it will sort the year, because years are index here
print(release_year_counts)
plt.figure(figsize=(10,6))
plt.scatter(release_year_counts.index,release_year_counts.values)
plt.title("Release year vs Number of Shows")
plt.xlabel("Release Year")
plt.ylabel("Number of Shows")
plt.tight_layout()
plt.savefig("ReleaseYear_VS_NumberOfShows.png")
plt.show()
```

```python
#top 10 countries which has most number of shows(horizontal graph)
country_counts=df["country"].value_counts().head(10)  #total count of country = total number of shows given by that country here
print(country_counts)
plt.figure(figsize=(10,5))
plt.barh(country_counts.index,country_counts.values,color="goldenrod")
plt.title("Top 10 Countries by Number of Shows")
plt.xlabel("Number of Shows")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig("10_countries_by_number_of_shows.png")
plt.show()
```

```python
#create subplot for movie vs tv-shows by year
content_by_year=df.groupby(["release_year","type"]).size()
content_by_year
#explanation of CODE:
# df.groupby(["release_year", "type"])
# Groups your data by both release_year and type.

# So it creates pairs like:
# (2019, Movie)
# (2020, TV Show)
# (2020, Movie)

# .size()
# Counts how many rows exist for each group (i.e., how many entries of each type per year).
# release_year  type
# 2019          Movie       300
#               TV Show     150
# 2020          Movie       320
#               TV Show     180

# .unstack()
# Converts the inner index (type) into columns.
# So now your DataFrame looks like this:
# | release\_year | Movie | TV Show |
# | ------------- | ----- | ------- |
# | 2019          | 300   | 150     |
# | 2020          | 320   | 180     |

# . .fillna(0)
# Fills any missing values with 0. For example, if in 2008 only Movies are present and no TV Shows, then TV Show will be NaN â†’ converted to 0.
```

```python
content_by_year=df.groupby(["release_year","type"]).size().unstack()
content_by_year
```

```python
content_by_year=df.groupby(["release_year","type"]).size().unstack().fillna(0)
content_by_year
```

```python
content_by_year=df.groupby(["release_year","type"]).size().unstack().fillna(0)

fig,ax=plt.subplots(1,2,figsize=(12,5))
plt.suptitle("Comparison of Movies/TV Shows released per Year")

#first subplot: movies
ax[0].plot(content_by_year.index,content_by_year["Movie"],color="steelblue")
ax[0].set_title("Movies Released Per Year")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Number of Movies Released")

#second subplot: TV Show
ax[1].plot(content_by_year.index,content_by_year["TV Show"],color="teal")
ax[1].set_title("TV Shows Released Per Year")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("Number of TV Shows Released")
plt.tight_layout()
plt.savefig("Movie_TVShows_released_per_year")
plt.show()
```

```python
#Extra Info:
# plt.tight_layout(): Automatically adjusts the subplot layout within the full figure area (rect=[0, 0, 1, 1] is assumed)
# If you add a suptitle(), it might overlap or get cut off, because tight_layout() doesn't account for extra space needed at the top.

# plt.tight_layout(rect=[0, 0, 1, 0.95])
# Tells Matplotlib: "Leave space at the top (i.e., only adjust subplots in 95% of the height), so the title fits."
# rect=[left, bottom, right, top] â†’ here top=0.95 leaves 5% space at the top for suptitle.
```
