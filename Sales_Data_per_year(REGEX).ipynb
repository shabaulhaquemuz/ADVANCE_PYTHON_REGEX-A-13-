```python
#order date me se month nikalna ha,,, unique month dekhna hai nikal aya ni... check datatype of month... make new column calleed as sales = quantity ordered*price of each
```

```python
copied_df["Order Date"]=pd.to_datetime(copied_df["Order Date"],errors="coerce")
copied_df = copied_df.dropna(subset=["Order Date"])
copied_df["Month"]=copied_df["Order Date"].dt.month
#copied_df["Month"]=copied_df["Order Date"].dt.month
```

```python
copied_df["Quantity Ordered"] = pd.to_numeric(copied_df["Quantity Ordered"], errors='coerce')
copied_df["Price Each"] = pd.to_numeric(copied_df["Price Each"], errors='coerce')
```

```python
#copied_df[['Year', 'Month', 'Day']] = copied_df['Order Date'].str.split('-', expand=True)
copied_df["sales"]=copied_df["Quantity Ordered"] * copied_df["Price Each"]
copied_df
```

```python
#1. . Total Sales
copied_df["sales"].sum()
```

```python
# 2. sales by month
copied_df.groupby("Month")["sales"].sum()
```

```python
# 3. Best-Selling Product
best_product=copied_df.groupby("Product")["sales"].sum().sort_values(ascending=False)
print(best_product.head(1))
```

```python
#4.Show unique products sold
print(copied_df["Product"].unique())
```

```python
#5.Count how many orders per product
print(copied_df.groupby("Product")["Quantity Ordered"].sum())
```

```python
#6.Total revenue per product
print(copied_df.groupby("Product")["sales"].sum())
```
